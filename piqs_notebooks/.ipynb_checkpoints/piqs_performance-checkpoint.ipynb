## Performance of PIQS 

Here we compare the performance of three different methods performing two tasks, building the matrix for the dynamics and solving the relative differential equation in time: 

1) The 'PIQS method'uses the Dicke space. We build the Liouvillian and integrate in time the master equation with QuTip's 'mesolve()'.

2) The 'PIM method' uses the reduced Dicke space, i.e. only the diagonal elements of the Dicke space. It applies only in a subset of cases (intial state and Hamiltonian diagonal in the Dicke space). We build the dyanmical matrix and integrate in time the rate equation with an ad-hoc solver.

3) The 'uncoupled method' uses the full space with uncoupled spins. We build the Liouvillian and integrate in time the master equation with QuTip's 'mesolve()'.

The algorithms are solved for a list of $N$ two-level systems (TLSs). The data is saved in .npz files as dicitionaries. The data is plotted using linear and semi-logarithmic scaling.

```python
import matplotlib.pyplot as plt
from qutip import *
from piqs import *
import numpy as np
from piqs import num_dicke_states as nds
from dicke_pim import *
import numpy.polynomial.polynomial as poly
from random import random as rnd
from time import clock
from piqs import excited as excited_dicke
from dicke_pim import Pim
```

## 1. Time scaling: Matrix load and integration time

```python
## Build "N TLSs" Lists for the 3 different methods

# Fine grained list for small N (list1)
# resolution on number of qubits (if == 1 every n is calculated)
factor_n = 1
# maximum number of qubits is given by nmax times factor_n
nmax = 10
# list of qubits
ntls_list1 = [i * factor_n for i in range(1,nmax+1)]

# Coarse grained list for large N (list2)
# resolution on number of qubits (if == 1 every n is calculated)
factor_n = 10
# maximum number of qubits is given by nmax times factor_n
nmax = 10
# list of qubits
ntls_list2 = [i * factor_n for i in range(1,nmax+1)]

ntls_list = ntls_list1 + ntls_list2
ntls_list = np.unique(ntls_list)

# eliminate N=1 from list
ntls_list = ntls_list[1:]

# 1 - PIQS method list 
ntls_piqs_list = ntls_list

# 2 - PIM method list 
ntls_pim_list = ntls_list

# 3 - UNCOUPLED method list 
nmax_uncoupled = 10
factor_n = 1
ntls_uncoupled_list = [i * factor_n for i in range(1,nmax_uncoupled+1)]
ntls_uncoupled_list = ntls_uncoupled_list[1:]
# list with number of Dicke states
nds_list = [nds(n) for n in ntls_list ]

print("ntls_list", ntls_list)
print("ntls_piqs_list", ntls_piqs_list)
print("ntls_pim_list", ntls_pim_list)
print("ntls_uncoupled_list", ntls_uncoupled_list)
print("dicke_states_list", nds_list)
```

```python
## Dynamical coefficients
gE = 0.1
gD = 0.1
gP = 0.1
gCE = 0.01
gCD = 0.01
gCP = 0.01
w0 = 1
wi = 0.1
## Set the lists that will be used to save the data
# 1 - Dicke basis lists (PIQS)
liouv_time_list = []
liouv_list = []
mesolve_time_list = []
jzt_list = []
jpjmt_list = []
# 2 - Reduced Dicke basis lists (PIM)
matrix_pim_time_list = []
solve_pim_time_list = []
jzt_pim_list = []
jpjmt_pim_list = []
# 3 - Uncoupled basis comparison (Uncoupled)
liouv_uncoupled_time_list = []
liouv_uncoupled_list = []
mesolve_uncoupled_time_list = []
jzt_uncoupled_list = []
jpjmt_uncoupled_list = []
# integration time
nt = 1000
```

```python
## Set matrix load Conditions
# method 1
load_piqs = False
# method 2
load_pim = False
# method 3
load_uncoupled = True

## Set time Integration Conditions
time_integration = True
# method 1
integrate_piqs = False
# method 2
integrate_pim = False
# method 3
integrate_uncoupled = True

for n in ntls_list:
    N = int(n)
    ntls = N
    [jx, jy, jz, jp, jm] = jspin(n)
    #spin hamiltonian
    w0 = 1
    h0 = w0 * jz
    hint = wi * jx
    h = h0 #+ hint
    #set initial conditions for spins
    system = Dicke(#hamiltonian = h, 
                   N = N,
                   emission = gE,
                   pumping = gP, 
                   dephasing = gD, 
                   collective_emission = gCE, 
                   collective_pumping = gCP, 
                   collective_dephasing = gCD)
    systempim = Pim(N = N, 
                   emission = gE,
                   pumping = gP, 
                   dephasing = gD, 
                   collective_emission = gCE, 
                   collective_pumping = gCP, 
                   collective_dephasing = gCD)
    if load_piqs == True and n in ntls_piqs_list:        
    # 1 - PIQS (Dicke space, calculate Liouvillian)
        clock_t0 = clock()
        liouv = system.liouvillian()
        #stop timer L
        clock_tf = clock()
        dt_clock = clock_tf - clock_t0
        liouv_time_list.append(dt_clock)
        liouv_list.append(liouv)
        print(" N = ", N, " PIQS - Liouvillian loaded, dt = ",dt_clock)
        clock_t0 = clock()
    if load_pim == True and n in ntls_pim_list:                
        # 2 - PIM - matrix load (reduced dyanmical matrix, diagonal elements in Dicke space)
        clock_t0 = clock()
        L = systempim.generate_matrix()
        clock_tf = clock()
        dt_clock = clock_tf - clock_t0
        print(" N = ", N, " PIM - Diagonal solver matrix loaded, dt = ",dt_clock)
        matrix_pim_time_list.append(dt_clock)
    if load_uncoupled == True and n in ntls_uncoupled_list:    
    # 3 - Uncoupled spin hamiltonian (full 4^N space, calculate Liouvillian)
        [jx_uncoupled, jy_uncoupled, jz_uncoupled, jp_uncoupled, jm_uncoupled] = jspin(n,basis="uncoupled")
        h0_uncoupled = w0 * jz_uncoupled
        hint_uncoupled = wi * jx_uncoupled
        h_uncoupled = h0_uncoupled #+ hint_uncoupled
        #set initial conditions for spins
        cops_uncoupled = collapse_uncoupled(N = N, 
                       emission = gE,
                       pumping = gP, 
                       dephasing = gD, 
                       collective_emission = gCE, 
                       collective_pumping = gCP, 
                       collective_dephasing = gCD) 
        #start timer L
        clock_t0 = clock()
        liouv_uncoupled = liouvillian(h_uncoupled, cops_uncoupled)
        #stop timer L
        clock_tf = clock()
        dt_clock = clock_tf - clock_t0
        liouv_uncoupled_time_list.append(dt_clock)
        liouv_uncoupled_list.append(liouv_uncoupled)
        print(" N = ", N, " Uncoupled - Liouvillian loaded, dt = ",dt_clock)
    # Time integration
    if time_integration == True:
        # General paralmeters
        clock_t0 = clock()
        td0 = np.log(N)/(N*gCE)
        tmax = max(4 * (td0),4) 
        t = np.linspace(0, tmax, nt)
        excited_state = dicke(N, N/2., N/2.)
        rho0 = excited_state
        # 1 - PIQS - Start timer (use 'mesolve()'in Dicke basis Liouvillian space)
        if integrate_piqs == True and n in ntls_piqs_list:        
            result = mesolve(liouv, rho0, t, [], e_ops = [jz, jp*jm], 
                             options = Options(store_states=True))
            rhot = result.states
            jzt = result.expect[0]
            jpjmt = result.expect[1]
            # stop timer
            clock_tf = clock()
            dt_clock = clock_tf - clock_t0
            print(" N = ", N, " PIQS - mesolve solved, dt = ",dt_clock)
            mesolve_time_list.append(dt_clock)
            jzt_list.append(jzt)    
            jpjmt_list.append(jpjmt) 
        # 2 - PIM - time integration (diagonal solver in reduced space)
        if integrate_pim == True and n in ntls_pim_list:        
            clock_t0 = clock()
            rhot_pim = systempim.solve(rho0, t)
            rho_diag = np.array([np.diag(np.real(i.full())) for i in rhot_pim.states], dtype=float)
            clock_tf = clock()
            dt_clock = clock_tf - clock_t0
            solve_pim_time_list.append(dt_clock)
            jzt_pim = expect(rhot_pim.states, jz)
            jzt_pim_list.append(jzt_pim)
            print(" N = ", N, " PIM - Diagonal solver time integrated, dt = ",dt_clock)
        # 3 - Uncoupled basis - time integration (use 'mesolve()' in full 4^N Liouvillian space)
        if integrate_uncoupled == True and n in ntls_uncoupled_list:            
            clock_t0 = clock()
            td0 = np.log(N)/(N*gCE)
            tmax = max(4 * (td0),4) 
            t = np.linspace(0, tmax, nt)
            #start timer mesolve
            rho0_uncoupled = excited_dicke(n, basis="uncoupled")
            result_uncoupled = mesolve(liouv_uncoupled, 
                                       rho0_uncoupled, t, [], 
                                       e_ops = [jz_uncoupled, jp_uncoupled*jm_uncoupled],
                                       options = Options(store_states=True))
            rhot_uncoupled = result_uncoupled.states
            jzt_uncoupled = result_uncoupled.expect[0]
            jpjmt_uncoupled = result_uncoupled.expect[1]
            #stop timer
            clock_tf = clock()
            dt_clock = clock_tf - clock_t0
            print(" N = ", N, " Uncoupled - mesolve solved, dt = ",dt_clock)
            mesolve_uncoupled_time_list.append(dt_clock)
            jzt_uncoupled_list.append(jzt_uncoupled) 
```

```python
# save data to file
save_file = True
if save_file == True:
    # 1 - Dicke basis (PIQS)
    if load_piqs == True:
        file_name = str("data/ntls_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), ntls_list)
        file_name = str("data/liouv_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), liouv_time_list)
        print("load_piqs saved")
    if integrate_piqs == True:
        file_name = str("data/mesolve_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), mesolve_time_list)
        file_name = str("data/jzt_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), jzt_list)
        print("integrate_piqs saved")
    # 2 - Diagonal Dicke basis (PIM)
    if load_pim == True:
        file_name = str("data/ntls_pim_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), ntls_pim_list)
        file_name = str("data/matrix_pim_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), matrix_pim_time_list)
        print("load_pim saved")
    if integrate_pim == True:        
        file_name = str("data/solve_pim_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), solve_pim_time_list)
        file_name = str("data/jzt_pim_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), jzt_pim_list)
        print("integrate_pim saved")
    # 3 - Uncoupled basis (uncoupled)
    if load_uncoupled == True:
        file_name = str("data/ntls_uncoupled_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), ntls_uncoupled_list)
        file_name = str("data/liouv_uncoupled_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), liouv_uncoupled_time_list)
        print("load_uncoupled saved")
    if integrate_uncoupled == True:        
        file_name = str("data/mesolve_uncoupled_time_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), mesolve_uncoupled_time_list)
        file_name = str("data/jzt_uncoupled_list_Nmax{}.npz".format(N))
        np.savez('{}'.format(file_name), jzt_uncoupled_list)
        print("integrate_uncoupled saved")
```

```python
# uncoupled basis comparison
liouv_uncoupled_time_list = []
liouv_uncoupled_list = []
mesolve_uncoupled_time_list = []
rhot_uncoupled_list = []
jzt_uncoupled_list = []
jpjmt_uncoupled_list = []
nt = 1000
random_coefficients = False
#start cycle
nmax_uncoupled = 9
factor_n = 1
ntls_uncoupled_list = [i * factor_n for i in range(1,nmax_uncoupled+1)]
print(ntls_list)
for n in ntls_uncoupled_list:
    N = int(n)
    ntls = N

    #coefficients
    if random_coefficients == True:
        gE = rnd()*1#np.sqrt(2)
        gD = rnd()*1#0.444
        gP = rnd()*1#0.1
        gCE = rnd()*1
        gCD = rnd()*1#np.pi
        gCP = rnd()*1#1.3    
        wi = rnd()*0.1
    else:
        gE = 1
        gD = 1
        gP = 1
        gCE = 0.1
        gCD = 0.1
        gCP = 0.1    
        wi = 0.3
    # uncoupled spin hamiltonian
    w0 = 1
    h0 = w0 * jz
    hint = wi * jx
    h = h0 #+ hint
    #set initial conditions for spins
    [jx, jy, jz, jp, jm] = jspin(n, basis="uncoupled")
    cops_tls = collapse_uncoupled(N = N, 
                   emission = gE,
                   pumping = gP, 
                   dephasing = gD, 
                   collective_emission = gCE, 
                   collective_pumping = gCP, 
                   collective_dephasing = gCD) 
    #start timer L
    clock_t0 = clock()
    liouv_uncoupled = liouvillian(h, cops_tls)
    #stop timer L
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    liouv_uncoupled_time_list.append(dt_clock)
    liouv_uncoupled_list.append(liouv_uncoupled)
    print(N, " Liouvillian loaded, dt = ",dt_clock)
    clock_t0 = clock()
    td0 = np.log(N)/(N*gCE)
    tmax = max(4 * (td0),4) 
    t = np.linspace(0, tmax, nt)
    #start timer mesolve
    excited_uncoupled = excited_dicke(n, basis="uncoupled")
    rho0 = excited_uncoupled
    result_uncoupled = mesolve(liouv_uncoupled, rho0, t, [], e_ops = [jz, jp*jm], options = Options(store_states=True))
    rhot_uncoupled = result_uncoupled.states
    jzt_uncoupled = result_uncoupled.expect[0]
    jpjmt_uncoupled = result_uncoupled.expect[1]
    #stop timer
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    print(N, " mesolve solved, dt = ",dt_clock)
    mesolve_uncoupled_time_list.append(dt_clock)
    rhot_uncoupled_list.append(rhot_uncoupled)
    jzt_list.append(jzt_uncoupled) 
```

### Liouvillian spectrum and steady state time

```python
nmax = 10
factor_n = 10
ntls_list = [i * factor_n for i in range(1,nmax+1)]
nds_list = [nds(n) for n in ntls_list ]
print(nds_list)
print(ntls_list)
steady_time_list = []
nt = 101
random_coefficients = False
#start cycle 
for n in ntls_list:
    N = int(n)
    ntls = N
    [jx, jy, jz, jp, jm] = jspin(n)
    #coefficients
    gE = 1
    gD = 1
    gP = 1
    gCE = 0.1
    gCD = 0.1
    gCP = 0.1    
    wi = 0.1
    #spin hamiltonian
    w0 = 1
    h0 = w0 * jz
    hint = wi * jx
    h = h0 + hint
    #set initial conditions for spins
    system = Dicke(hamiltonian = h, N = N,
                   loss = gE,
                   pumping = gP,
                   dephasing = gD, emission = gCE,
                   collective_pumping = gCP,
                   collective_dephasing = gCD)
    #start timer L
    clock_t0 = clock()
    liouv = system.liouvillian()
    #stop timer L
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    liouv_time_list.append(dt_clock)
    print(N, "D loaded, dt = ",dt_clock)
    clock_t0 = clock()
    rho_ss = steadystate(liouv)
    #stop timer L
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    steady_time_list.append(dt_clock)

    #stop timer L
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    steady_time_list.append(dt_clock)
    print(N, "rhoss calculated, dt = ",dt_clock)
```

### Eigenvalues time

```python
nmax = 10
factor_n = 10
ntls_list = [i * factor_n for i in range(1,nmax+1)]
nds_list = [nds(n) for n in ntls_list ]
print(nds_list)
print(ntls_list)
eig_time_list = []
gE = 1
gD = 1
gP = 1
gCE = 0.1
gCD = 0.1
gCP = 0.1    
wi = 0.1
#spin hamiltonian
w0 = 1
ii = 0
#start cycle 
for n in ntls_list:
    N = int(n)
    ntls = N
    liouv = liouv_list[ii]
    ii = ii+1
    clock_t0 = clock()
    eigval, eigvec = liouv.eigenstates()
    #stop timer L
    clock_tf = clock()
    dt_clock = clock_tf - clock_t0
    eig_time_list.append(dt_clock)
    print(N, "eig calculated, dt = ",dt_clock)
```
